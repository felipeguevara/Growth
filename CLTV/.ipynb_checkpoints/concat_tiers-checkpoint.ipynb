```python
import pandas as pd

def concat_csv(file_list):
    df = pd.concat([pd.read_csv(f) for f in file_list])
    return df
```

```python
cities = ['BOG','BAQ','MDE','SPO','BHZ','CWB','VCP','CMX']
```

# Aca se crea el csv conjunto

```python
file_list = [f'{city}_SAC_tiers.csv' for city in cities]
```

```python
df = concat_csv(file_list)
```

```python
df.to_csv(f"ALL_SAC_tiers.csv",index=False)
```

# Aca se crea el Excel conjunto

```python
# Set the list of cities and filenames
file_list = [f"CLTV_{city}.xlsx" for city in cities]

# Create an empty dictionary to store the dataframes
dataframes = {}

# Loop through each file and read the data into a dataframe
for file in file_list:
    city = file.split("_")[1].split(".")[0]  # Extract the city name from the filename
    df = pd.read_excel(file)
    dataframes[city] = df
```

```python
dataframes['BOG']
```

```python
# Write the dataframes to a new XLSX file, with each city as a sheet
with pd.ExcelWriter("CLTV_ALL.xlsx") as writer:
    for city, df in dataframes.items():
        df.to_excel(writer, sheet_name=city, index=False)
```
